```python
import re

def cleanupData(df):
#     df = pd.DataFrame(df)
    df.columns = ['DatePulled', 'CustCode', 'EmployeeIndex', 'CustCountry', 'CustGender', 'CustAge', 'CustStartDate', 'NewCustIndex', 'CustSeniority', 'CustDuringNotEndMonth', 'CustLastDatePrimary',
                 'CustType', 'CustRelationType', 'ResidenceIndex', 'ForeignerIndex', 'EmployeeSpouse', 'CustChannelJoin', 'Deceased', 'AddressType', 'CustProvinceCode', 'CustProvince', 'CustActivity',
                 'CustGrossIncome', 'CustSegment', 'SavingsAcct', 'Guarantees', 'CurrentAcct', 'DerivAcct', 'PayrollAcct', 'JuniorAcct', 'SpecialAcct1', 'SpecialAcct2',
                 'SpecialAcct3Plus', 'STDeposits', 'MTDeposits', 'LTDeposits', 'eAcct', 'Funds', 'Mortgage', 'Pensions', 'Loans', 'Taxes', 'CreditCard', 'Securities', 'HomeAcct', 'Payroll', 'Pensions', 'DirectDebit']
    
    df2 = df.iloc[:,1:]
    del df2['AddressType']
    del df2['CustLastDatePrimary']
    del df2['CustStartDate']
    df = df[~df['CustAge'].isin(['NA', ' NA', '  NA', '   NA', '    NA', '     NA', '      NA', '       NA'])]
    df = df[~df['CustSeniority'].isin(['NA', ' NA', '  NA', '   NA', '    NA', '     NA', '      NA', '       NA'])]
    df['CustAge'] = pd.to_numeric(df['CustAge'], errors='coerce')
    df['CustSeniority'] = pd.to_numeric(df['CustSeniority'])
    df = df[df['CustSeniority'] > 0]
    
    return df
    
    
    
```

```python
def getdiff(df1, df2):
    df1.sort_values('CustCode', inplace=True)
    df2.sort_values('CustCode', inplace=True)
    
    df1 = df1.iloc[:,1:]
    df2 = df2.iloc[:,1:]
    
    df1['NumberProducts'] = df1.iloc[:,23:].sum(axis=1)
    df2['NumberProducts'] = df2.iloc[:,23:].sum(axis=1)
    
    df_comb = df1.merge(df2, on='CustCode')
    
    df_diff = df_comb[df_comb["NumberProducts_x"] != df_comb["NumberProducts_y"] ]
    df_diff['ChangeProd'] = df_comb["NumberProducts_y"] - df_comb["NumberProducts_x"]
    
    df_same = df_comb[df_comb["NumberProducts_x"] == df_comb["NumberProducts_y"] ].sample(50000)
    df_same['ChangeProd'] = 0
#     df = df.append(df_diff)
#     df = df.append(df_same)\
    df_comb = df_diff.append(df_same)
    
    return df_comb
    
```

```python
def plot_confusion_matrix(cm,title='Confusion matrix', cmap=plt.cm.Reds):
    plt.imshow(cm, interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def train_score(classifier,x,y):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=4444)
    ytrain=np.ravel(ytrain)
    clf = classifier.fit(xtrain, ytrain)
    # accuracy for test & train:
    train_acc=clf.score(xtrain, ytrain)
    test_acc=clf.score(xtest,ytest)
    print("Training Data Accuracy: %0.2f" %(train_acc))
    print("Test Data Accuracy:     %0.2f" %(test_acc))
    
    y_true = ytest
    y_pred = clf.predict(xtest)


    conf = confusion_matrix(y_true, y_pred)
    precision = (conf[0, 0] / (conf[0, 0] + conf[1, 0]))
    recall = (conf[0, 0] / (conf[0, 0] + conf[0, 1]))
    F1 = 2 * (precision * recall) / (precision + recall)
    
    print(conf)

    print ('\n')
    print ("Precision:          %0.2f" %precision)
    print ("Recall:             %0.2f"%recall )
    print ("F1:                 %0.2f"%F1 )
    
    cm=confusion_matrix(y_true, y_pred, labels=None)
    
    plt.figure()
    plot_confusion_matrix(cm)
```
