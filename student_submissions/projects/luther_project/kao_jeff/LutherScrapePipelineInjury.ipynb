##### Scraping Injury Tables from http://www.prosportstransactions.com/

```python
import requests
import re
import pandas as pd
import string
from bs4 import BeautifulSoup
```

```python
#URL:  http://www.prosportstransactions.com/basketball/Search/SearchResults.php?Player=&Team=&BeginDate=1980-07-01&EndDate=&ILChkBx=yes&InjuriesChkBx=yes&PersonalChkBx=yes&Submit=Search&start=0
base_url = 'http://www.prosportstransactions.com/basketball/Search/SearchResults.php?Player=&Team=&BeginDate=1980-07-01&EndDate=&ILChkBx=yes&InjuriesChkBx=yes&PersonalChkBx=yes&Submit=Search&start={}'
```

```python
def get_injury_links():
    returnlist = []
    for i in range(1265,2102):
        returnlist.append(base_url.format(i*25))
    return returnlist
```

```python
def parse_injury_table(page_text):
    soup = BeautifulSoup(page_text, "lxml")
    # find links
    identifier = 'datatable center'
    table = soup.find('table', { "class" : identifier})
    rows = table.find_all('tr')
    # initialize the LIST for injury stats (we're going to return a list of lists)
    table_rows = []
    df = pd.DataFrame()
    #first row headers
    for row in rows[1:]:
        cell_list = [] # temp new list
        tds = row.find_all('td')
        for td in tds:
            cell_list.append(td.text)
        table_rows.append(cell_list)
    return table_rows
```

```python
##### NEED TO WORK ON THIS #####
injuries_list = []
paths = get_injury_links()
injury_fails = []
print('=== DOWNLOADING INJURIES ===')
for path in paths:
    fullpath = base_url + path
    response = requests.get(fullpath)
    if response.status_code == 200:
        temp_table = parse_injury_table(response.text)
        injuries_list.append(temp_table)
        print('Injuries Downloaded Successfully: {}'.format(fullpath))
    else:
        print('Injuries Download Failed: {}'.format(fullpath))
        injury_fails.append(fullpath)
print('=== DOWNLOADING INJURIES COMPLETED! ===')
injuries_list
```

```python
flat_list = [item for sublist in injuries_list for item in sublist]
```

```python
injuries_df = pd.DataFrame(flat_list)
```

```python
injuries_df.columns=['date','team','acquired','relinquished','notes']
```

```python
injuries_df.to_csv('injuries_interim4.csv')
```

```python
#season year -- cut the date at July of the year before
injuries_df['date'] = pd.to_datetime(injuries_df['date'], infer_datetime_format=True)
def get_season_year(datetime):
    year = datetime.year
    if datetime.month >= 7:
        year += 1
    return year
injuries_df['season_year'] = injuries_df['date'].apply(get_season_year)
```
